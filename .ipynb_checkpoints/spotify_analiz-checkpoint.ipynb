```python
import spotipy #Spotipy, Spotify'ın sunduğu tüm müzik verilerine ve özelliklerine erişmenizi sağlayan, Python dili için geliştirilmiş hafif bir kütüphanedi
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def sanatci_analiz_et_guncel(sanatci_adi):
    # 1. Sanatçıyı bul
    results = sp.search(q=sanatci_adi, limit=1, type='artist')
    if not results['artists']['items']:
        return "Sanatçı bulunamadı."
    
    artist = results['artists']['items'][0]
    artist_uri = artist['uri']
    
    # 2. En popüler 10 şarkısını al
    top_tracks = sp.artist_top_tracks(artist_uri)
    
    liste = []
    for track in top_tracks['tracks']:
        liste.append({
            'Şarkı Adı': track['name'],
            'Popülerlik': track['popularity'],
            'Albüm': track['album']['name'],
            'Yayın Yılı': track['album']['release_date'][:4], # Sadece yılı al
            'Süre (dk)': round(track['duration_ms'] / 60000, 2),
            'Albüm Tipi': track['album']['album_type']
        })
    
    df = pd.DataFrame(liste)
    # Popülerliğe göre sıralayalım
    return df.sort_values(by='Popülerlik', ascending=False)

# Test edelim
df = sanatci_analiz_et_guncel("Sezen Aksu")
df
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Popülerlik', y='Şarkı Adı', palette='viridis')

plt.title(f"Sezen Aksu - En Popüler 10 Şarkı Analizi")
plt.xlabel("Popülerlik Puanı (0-100)")
plt.ylabel("Şarkı")
plt.show()
```

```python
# Sanatçının tür verilerini çekelim
results = sp.search(q="Sezen Aksu", limit=1, type='artist')
artist = results['artists']['items'][0]
turler = artist['genres']

print(f"Sanatçı Türleri: {turler}")

# Eğer birden fazla tür varsa (ki genellikle öyledir) bunları analiz edebiliriz
if turler:
    plt.figure(figsize=(8, 4))
    # Basit bir analiz: Türlerin popülerlik içinde kapladığı alan gibi düşünebiliriz
    sns.countplot(y=turler, palette='magma', hue=turler, legend=False)
    plt.title("Sanatçının Dahil Olduğu Müzik Türleri")
    plt.show()
```
